```python
# 알고리즘 개선하기 (1)
# - 2의 배수는 소수가 아님 → 2의 배수인지 확인
# - 2의 배수가 아니면 3부터 N-1까지의 홀수에 대해서만 나누어지는지 확인

def is_prime_number(n):
    if n == 2: return True
    # 2의 배수는 여기서 거름
    if n == 1 or n%2 == 0: return False
    # 2의 배수는 건너뛰고 확인
    for i in range(3, n, 2):
        if n % i == 0:
            return False
    return True
print(is_prime_number(5))
```

```python
# 알고리즘 개선하기 (2)
# - 약수를 구성하는 수는 N/2 이하의 수와 N/2 이상의 수로 구성됨
# → N/2 이하의 수로 나누어지지 않으면 N/2 이상의 수로도 나누어지지 않음
def is_prime_number(n):
    if n == 2: return True
    if n == 1 or n%2 == 0: return False
    for i in range(3, n//2, 2):
        if n % i == 0:
            return False
    return True
r = int(input())
print(is_prime_number(r))
```

# 투 포인터, Two Pointer

```python
# BOJ 11728번(배열 합치기)
# - 정렬되어 있는 배열을 하나의 정렬된 배열로 합치기
lst = []
n, m = map(int, input().split())

pa = list(map(int, input().split()))
pb = list(map(int, input().split()))
x=0
y=0
        
while x < n and y < m:
    if pa[x] < pb[y]:
        lst.append(pa[x])
        x += 1
    elif pa[x] >= pb[y]:
        lst.append(pb[y])
        y += 1
lst += pa[x:]
lst += pb[y:]
for i in lst:
    print(i, end=" ")
```

```python
# BOJ 2003번 (수들의 합 2)
# - 배열의 연속된 요소들의 합이 주어진 값이 되는 경우의 수 구하기
# - 이전 구간합으로부터 start와 end를 움직여서 구간 합 update하기

n, m = map(int, input().split())
numlst = list(map(int, input().split()))

answer = 0
start = 0
end = 0
sum = 0

answer = 0
start = end = 0
sum = numlst[0]
while end < n:
    if sum == m:
        answer += 1
    if sum <= m:
        end +=1
        if end<n:
            sum += numlst[end]
    else:
        sum -= numlst[start]
        start += 1
print(answer)
```

# 진법변환

```python
# 정수 N을 받아서  base에 대한 진법의 수로 변환

def convert_base(N, base):
    # N을 base로 나눈 다음 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    # 나눈 나머지를 진법에 해당하는 숫자를 변환하여 리스트에 저장한다.
    # N: 몫을 저장
    # new_list: 나머지를 저장하는 리스트
    new_list = []
    digits='0123456789ABCDEFGHIGKLMNOPQRSTUVWXYZ'
    while N != 0:
        new_list.append(digits[N % base])
        N //= base
    return new_list

print(convert_base(25,2))
```

```python

def convert_base(N, base):
    # N을 base로 나눈 다음 몫을 몫이 0이 아닌 동안 계속 N으로 나눈다.
    # 나눈 나머지를 진법에 해당하는 숫자를 변환하여 리스트에 저장한다.
    # N: 몫을 저장
    # new_list: 나머지를 저장하는 리스트
    new_number = ''
    digits='0123456789ABCDEFGHIGKLMNOPQRSTUVWXYZ'
    while N != 0:
        new_number = digits[N % base] + new_number
        N //= base
    return new_number

print(convert_base(25,2))
```
