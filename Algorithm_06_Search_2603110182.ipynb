```python
# recursive call을 활용한 binary search
def binary_search_recursive(A, Key, First, Last):
    if (First > Last):
        # 찾는 키가 A에 없는 경우(탐색할 항목이 더이상 없다)
        return -1

    # 중앙값 찾기
    middle = (First + Last) // 2

    # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
    if Key == A[middle]:
        return middle
    # 탐색영역을 왼쪽으로 설정
    elif Key < A[middle]:
        return binary_search_recursive(A, Key, First, middle-1)
    # 탐색영역을 오른쪽으로 설정
    else:
        return binary_search_recursive(A, Key, middle+1, Last)
```

```python
# while loop를 활용한 binary search
def binary_search(A, Key):
    # 탐색영역의 시작
    Last = 0
    # 탐색영역의 마지막
    First = len(A) - 1
    
    while Last <= First:
        # 중앙값 찾기
        middle = (Last + First) // 2
        # 중앙에 있는 값과 비교하여 같으면 해당 index를 return
        if A[middle] == Key:
            return middle
        # 작으면 -> 왼쪽영역을 탐색 Last값 변경
        elif A[middle] < Key:
            Last = middle + 1
        # 크면 -> 오른쪽 영역 탐색 First값 변경
        else:
            First = middle - 1
    # 찾는 data가 없으면 -1
    return -1
```

```python
data = [2, 6, 11, 13, 18, 20, 22, 27, 29, 30, 34, 38, 41, 42, 45, 47]

num = int(input('찾을 숫자:'))
index = binary_search_recursive(data, num, 0 ,len(data))
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 data {index+1}째 칸에 있습니다')


index = binary_search(data, num)
if index == -1:
    print('없는 숫자입니다.')
else:
    print(f'>>> 숫자 {num}은 data {index+1}째 칸에 있습니다')
```
