### 1.이진트리

```python
class TreeNode:

    def __init__(self, data, left, right):
        self.data = data
        self.left = left
        self.right = right
```

```python
# 바이너리 트리 생성 (맨 밑 리프노드부터 만드는 것이 편함)

d = TreeNode('D', None, None)
e = TreeNode('E', None, None)
b = TreeNode('B', d, e)

f = TreeNode('F', None, None)
c = TreeNode('C', f, None)

a = TreeNode('A', b, c)

my_binary_tree =  a
```

```python
# preorder (전위순회)

def preorder(node):
    print(node.data, end=' ') #루트노드
    if node.left: 
        preorder(node.left) #왼쪽노드
    if node.right: 
        preorder(node.right) #오른쪽노드

preorder(my_binary_tree)
```

```python
# inorder (중위순회)

def inorder(node):
    # 자식 노드가 없으면 리턴
    if node == None:
        return    
    
    inorder(node.left) #왼쪽노드
    print(node.data, end=' ') #루트노드
    inorder(node.right) #오른쪽노드

inorder(my_binary_tree)
```

```python
# postorder (후위순회)

def postorder(node):
    if node == None:
        return
    
    postorder(node.left) #왼쪽노드
    postorder(node.right) #오른쪽노드
    print(node.data, end=' ') #루트노드

postorder(my_binary_tree)
```

```python
# 딕셔너리로 구현한 이진트리구조

my_binary_tree_dict = {}
my_binary_tree_dict['A'] = ['B', 'C']
my_binary_tree_dict['B'] = ['D', 'E']
my_binary_tree_dict['C'] = ['F', None]
my_binary_tree_dict['D'] = [None, None]
my_binary_tree_dict['E'] = [None, None]
my_binary_tree_dict['F'] = [None, None]
```

```python
# 딕셔너리로 구현한 preorder (전위순회)

def preorder(node):
    print(node, end=' ') #루트노드
    if my_binary_tree_dict[node][0]: 
        preorder(my_binary_tree_dict[node][0]) #왼쪽노드
    if my_binary_tree_dict[node][1]: 
        preorder(my_binary_tree_dict[node][1]) #오른쪽노드

preorder('A')
```

```python
# 딕셔너리로 구현한 inorder (중위순회)

def inorder(node):
    # 자식 노드가 없으면 리턴
    if node == None:
        return    
    
    inorder(my_binary_tree_dict[node][0]) #왼쪽노드
    print(node, end=' ') #루트노드
    inorder(my_binary_tree_dict[node][1]) #오른쪽노드

inorder('A')
```

```python
# 딕셔너리로 구현한 postorder (후위순회)

def postorder(node):
    if node == None:
        return
    
    postorder(my_binary_tree_dict[node][0]) #왼쪽노드
    postorder(my_binary_tree_dict[node][1]) #오른쪽노드
    print(node, end=' ') #루트노드

postorder('A')
```

### Binary Search Tree

```python
class STreeNode:

    def __init__(self, key, left, right):
        self.key = key
        self.left = left
        self.right = right
```

```python
# BST 생성
t3 = STreeNode(3, None, None)
t12 = STreeNode(12, None, None)
t7 = STreeNode(7, t3, t12)
t27 = STreeNode(27, None, None)
t31 = STreeNode(31, t27, None)
t26 = STreeNode(26, None, t31)
t18 = STreeNode(18, t7, t26)
my_bst = t18
```

```python
def search_bst(n, key):
    if n == None:
        return None
    if key == n.key:
        return n
    if key < n.key:
        return search_bst(n.left, key)
    else:
        return search_bst(n.right, key)
    
result = search_bst(my_bst, 25)
if result:
    print(result.key)
else:
    print('없는 key입니다')
```

```python
def search_bst_iter(n, key):
    while n != None:
        if key == n.key:
            return n
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return None

result = search_bst_iter(my_bst, 31)
```

```python
### class STreeNode로 숫자를 순서대로 입력받아 binary search tree를 구성하기 : insert_bst() 함수 활용 (iteration으로 코딩)

def insert_bst(n, key):
    while n != None:
        parent = n
        if key == n.key:
            return False
        elif key < n.key:
            n = n.left
        else:
            n = n.right
    return parent

# preorder (선위순회)
def preorder(node):
    print(node.key, end=' ')
    if node.left: preorder(node.left)
    if node.right: preorder(node.right)

turn = int(input('몇 번 입력? : '))
for i in range(turn):
    input_num = int(input('숫자를 입력하세요 : '))
    result = insert_bst(my_bst, input_num)

    if result is False:
        print('이미 있는 값입니다.')
    elif input_num < result.key:
        result.left = STreeNode(input_num, None, None)
        print(f'{result.key}의 왼쪽에 {input_num}입력되었습니다.')
    else:
        result.right = STreeNode(input_num, None, None)
        print(f'{result.key}의 오른쪽에 {input_num}입력되었습니다.')
    preorder(my_bst)
    print()
    
print('[종료]')
```
